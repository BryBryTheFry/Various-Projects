```python
import mediapipe as mp
import cv2
import numpy as np
import uuid
import os
```

```python
mp_drawing = mp.solutions.drawing_utils
mp_hands = mp.solutions.hands
```

```python
counting = {
    1 : 0,
    3 : 1,
    7 : 2,
    8 : 3,
    15 : 3,
    16 : 4,
    31 : 4,
    32 : 5,
    
}
```

```python
cap = cv2.VideoCapture(0)

with mp_hands.Hands(min_detection_confidence=0.8, min_tracking_confidence=0.5) as hands: 
    while cap.isOpened():
        ret, frame = cap.read()
        
        # BGR 2 RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        
        # Flip on horizontal
        image = cv2.flip(image, 1)
        
        # Set flag
        image.flags.writeable = False
        
        # Detections
        results = hands.process(image)
        
        # Set flag to true
        image.flags.writeable = True
        
        # RGB 2 BGR
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        # Detections
        # print(results)
        
        # Rendering results
        if results.multi_hand_landmarks:
            keypoints = []
            for num, hand in enumerate(results.multi_hand_landmarks):
                
                for i in hand.landmark:
                    x = int(i.x * frame.shape[1])
                    y = int(i.y * frame.shape[0])
                    keypoints.append([x, y])
                
                handID = 32
                
                if ((keypoints[4][0] < keypoints[17][0]) and (keypoints[4][0] > keypoints[5][0])) or ((keypoints[4][0] > keypoints[17][0]) and (keypoints[4][0] < keypoints[5][0])) :
                    #print("Thumb")
                    handID -= 1
                    
                if ((keypoints[8][1] > keypoints[9][1]) and (keypoints[8][1] < keypoints[0][1])):
                    #print("Index")
                    handID -= 2
                    
                if ((keypoints[12][1] > keypoints[9][1]) and (keypoints[12][1] < keypoints[0][1])):
                    #print("Middle")
                    handID -= 4
                    
                if ((keypoints[16][1] > keypoints[9][1]) and (keypoints[16][1] < keypoints[0][1])):
                    #print("Ring")
                    handID -= 8
                    
                if ((keypoints[20][1] > keypoints[9][1]) and (keypoints[20][1] < keypoints[0][1])):
                    #print("Pinky")
                    handID -= 16
                    
                if handID in counting:
                    print(counting[handID])
                
                mp_drawing.draw_landmarks(image, hand, mp_hands.HAND_CONNECTIONS, 
                                        mp_drawing.DrawingSpec(color=(121, 22, 76), thickness=2, circle_radius=4),
                                        mp_drawing.DrawingSpec(color=(250, 44, 250), thickness=2, circle_radius=2),
                                         )
            
        
        cv2.imshow('Hand Tracking', image)
        
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
